```python
data_dir = '/content/drive/MyDrive/Dataset/Images/'
```

```python
# Initialize empty lists for features and labels
X = []
Y = []
# Loop through each label folder
for label in os.listdir(data_dir):
    label_dir = os.path.join(data_dir, label)
    if os.path.isdir(label_dir):
        # Loop through each file in the label folder
        for filename in os.listdir(label_dir):
            file_path = os.path.join(label_dir, filename)
            # Extract features and add to the dataset
            X.append(np.load(file_path))
            Y.append(label)
# Convert to numpy arrays
X = np.array(X)
Y = np.array(Y)
```

```python
# Z-score normalization
X_mean = X.mean()
X_std = X.std()
X = (X - X_mean) / X_std
```

```python
print(f"Shape of X: {X.shape}")
print(f"Shape of Y: {Y.shape}")
```

```python
X = X.astype(np.int32)  # Or np.int64 if needed
Y = Y.astype(np.int32)  # Or np.int64 if needed
np.save('/content/drive/MyDrive/Dataset/X.npy', X)
np.save('/content/drive/MyDrive/Dataset/Y.npy', Y)
```
