```python
import os
import pandas as pd
```

```python
folder_path = os.getcwd()
evaluation_file = r'IEMOCAP_full_release\Session2\dialog\EmoEvaluation\Ses02F_impro01.txt'
true_labels = []
with open(os.path.join(folder_path, evaluation_file)) as f:
    for line in f.readlines():
        for word in line.split():
            if word.startswith('Ses'):
                temp_tuple = (word, line.split()[line.split().index(word) + 1])
                true_labels.append(temp_tuple)
print(true_labels)
```

```python
# obtaining true labels for all files from Session 1
directory = os.path.join(folder_path, r'IEMOCAP_full_release\Session2\dialog\EmoEvaluation')
true_labels = []
for file in os.listdir(directory):
    if file.endswith('.txt'):
        with open(os.path.join(directory, file)) as f:
            for line in f.readlines():
                for word in line.split():
                    if word.startswith('Ses'):
                        temp_tuple = (word, line.split()[line.split().index(word) + 1])
                        true_labels.append(temp_tuple)
print(true_labels)








```

```python
for element in true_labels:
    print(element)
    print()
```

```python
true_labels_df = pd.DataFrame(true_labels, columns = ['file_section_name', 'true_label'])
true_labels_df
```

```python
true_labels_df.to_csv(os.path.join(folder_path, r'SER_project_true_labels/S5_true_labels.csv'))
```
